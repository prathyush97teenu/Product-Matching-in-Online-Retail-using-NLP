```python
import nltk
#nltk.download('punkt')
from nltk.tokenize import word_tokenize
import numpy as np
import pandas as pd
import warnings
import regex as re
#pip install gensim
```

```python
df_offers_training = pd.read_excel(r'C:\Users\prath\Desktop\offers_test_cleaned.xlsx')
data = df_offers_training.iloc[:, :]
A = data.loc[data.shop == 'zalando', :]
B = data.loc[data.shop == 'aboutyou', :]
```

```python
mylist1=[]
mylist2=[]
a_list = A.values.tolist()
b_list = B.values.tolist()
```

```python
#A = A.drop_duplicates(subset=['text'])
#B = B.drop_duplicates(subset=['text'])
```

```python
#A.nunique(axis=0)
#B.nunique(axis=0)
```

```python
for i in range(len(A)):
  temp = a_list[i][3]
  mylist1.append(temp)
for i in range(len(B)):
  temp = b_list[i][3]
  mylist2.append(temp)
```

```python
print(mylist2)
```

```python
sentences = mylist2
```

```python
# Tokenization of each document
tokenized_sentences = []
for s in sentences:
    tokenized_sentences.append(word_tokenize(s.lower()))
tokenized_sentences
```

```python
def cosine_similarity(u, v):
    return np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))
```

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
tagged_data = [TaggedDocument(d, [i]) for i, d in enumerate(tokenized_sentences)]
tagged_data
```

```python
## Train doc2vec model
model = Doc2Vec(tagged_data, vector_size = 20, window = 2, min_count = 1, epochs = 100)

'''
vector_size = Dimensionality of the feature vectors.
window = The maximum distance between the current and predicted word within a sentence.
min_count = Ignores all words with total frequency lower than this.
alpha = The initial learning rate.
'''
```

```python
test_doc = word_tokenize(mylist1[3].lower())
test_doc_vector = model.infer_vector(test_doc)
print(model.docvecs.most_similar(positive = [test_doc_vector]))
```

```python
matches_list_1 = []
for i in range(len(A)):
    test_doc = word_tokenize(mylist1[i])
    test_doc_vector = model.infer_vector(test_doc)
    matches_list_1.append([i, model.docvecs.most_similar(positive = [test_doc_vector])])
```

```python
matches_list_1[0:5]
```

```python
matches_list_2 = []
for i in range(len(matches_list_1)):
    matches_list_2.append([i, a_list[i][0], a_list[i][3], matches_list_1[i][1][0][0], matches_list_1[i][1][0][1]])
```

```python
matches_list_2
```

```python
df_matches = pd.DataFrame(matches_list_2, columns = ['zalando_index', 'zalando_offer_id', 'zalando_text', 'aboutyou_index','similarity'])
```
